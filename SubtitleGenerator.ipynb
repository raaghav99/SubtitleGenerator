```python
from datetime import timedelta
import os
import whisper

from deep_translator import GoogleTranslator
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
audio = whisper.load_audio('/content/drive/MyDrive/k.wav')
print('audio loaded')
```

```python
beam_size=5
best_of=2
temperature=(0.0,  0.2,  0.4)

decode_options = dict(language="ja", best_of=best_of, beam_size=beam_size, temperature=temperature)
transcribe_options = dict(task="transcribe", **decode_options)
```

```python
model = whisper.load_model("large", device="cuda")
print("Whisper model loaded.")
```

```python
 # Change this to your desired model

transcribe = model.transcribe(audio)
print("transcription complete")
```

```python

segments = transcribe['segments']
print('segments complete')
```

```python
print(segments)
```

```python
#translator = Translator()
srt = open("k.srt", "x")
for segment in segments:
    startTime = str(0)+str(timedelta(seconds=int(segment['start'])))+',000'
    endTime = str(0)+str(timedelta(seconds=int(segment['end'])))+',000'
   # print(segment['text'])
    text=GoogleTranslator(source='ja', target='en').translate(segment['text'])
    segmentId = segment['id']+1
    segment = f"{segmentId}\n{startTime} --> {endTime}\n{text[1:] if text[0] is ' ' else text}\n\n"

  #  srtFilename = os.path.join("SrtFiles", f"VIDEO_FILENAME.srt")
   # with open(srtFilename, 'a', encoding='utf-8') as srtFile:
    srt.write(segment)
srt.close()
    #text = segment['text']
print('done')
```
